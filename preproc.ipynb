```python
import pandas as pd
```

```python
orders = pd.read_csv('data/raw/olist_orders_dataset.csv')
```

Miramos la cantidad de ordenes por mes

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_month'] = orders['order_purchase_timestamp'].dt.to_period('M')
orders_per_month = orders.groupby('order_month').size().reset_index(name='orders_count')
```

```python
orders_per_month
```

El dataset es inconsistente antes del 2017 y despues de Agosto del 2018.

Filtramos por estas fechas y guardamos en ``data/proc``

```python
orders_filtered = orders[(orders['order_purchase_timestamp'].dt.year >= 2017) & (orders['order_purchase_timestamp'].dt.year <= 2018) & (orders['order_purchase_timestamp'].dt.month <= 8)]
orders_per_month_filtered = orders_filtered.groupby('order_month').size().reset_index(name='orders_count')
orders_per_month_filtered
```

```python
orders_filtered.drop(columns=['order_month'], inplace=True)
orders_filtered.to_csv('data/proc/orders_filtered.csv', index=False)
```

```python
valid_order_ids = set(orders_filtered['order_id'])
valid_order_ids
```

```python
order_payments = pd.read_csv('data/raw/olist_order_payments_dataset.csv')
order_items = pd.read_csv('data/raw/olist_order_items_dataset.csv')
order_reviews = pd.read_csv('data/raw/olist_order_reviews_dataset.csv')
```

```python
order_payments_df = order_payments[order_payments['order_id'].isin(valid_order_ids)]
order_items_df   = order_items[order_items['order_id'].isin(valid_order_ids)]
order_reviews_df = order_reviews[order_reviews['order_id'].isin(valid_order_ids)]
```

```python
order_payments_df.to_csv('data/proc/olist_order_payments_dataset.csv', index=False)
order_items_df.to_csv('data/proc/olist_order_items_dataset.csv', index=False)
order_reviews_df.to_csv('data/proc/olist_order_reviews_dataset.csv', index=False)
```
